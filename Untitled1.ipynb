````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Langkah 1: Baca citra
img = cv2.imread('kucing-photoaidcom-greyscale.jpg', cv2.IMREAD_GRAYSCALE)

# Tambahkan pemeriksaan di sini
if img is None:
    print("Error: Gagal memuat citra. Pastikan nama file benar dan file tidak rusak.")
else:
    # Langkah 2: Terapkan transformasi
    negative_img = 255 - img

    c = 255 / np.log(1 + np.max(img))
    logarithmic_img = c * np.log(1 + img)
    logarithmic_img = np.uint8(logarithmic_img)  # Konversi ke uint8

    # Langkah 3: Tampilkan citra
    plt.figure(figsize=(15, 5))

    plt.subplot(131)
    plt.imshow(img, cmap='gray')
    plt.title('Citra Asli')

    plt.subplot(132)
    plt.imshow(negative_img, cmap='gray')
    plt.title('Citra Negatif')

    plt.subplot(133)
    plt.imshow(logarithmic_img, cmap='gray')
    plt.title('Citra Logaritmik')

    plt.show()

    # Langkah 4: Pembuatan Histogram
    plt.figure(figsize=(15, 5))

    plt.subplot(131)
    plt.hist(img.flatten(), 256, range=[0, 256], color='r')
    plt.title('Histogram Asli')

    plt.subplot(132)
    plt.hist(negative_img.flatten(), 256, range=[0, 256], color='r')
    plt.title('Histogram Negatif')

    plt.subplot(133)
    plt.hist(logarithmic_img.flatten(), 256, range=[0, 256], color='r')
    plt.title('Histogram Logaritmik')

    plt.show()
````

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Langkah 1: Baca citra
img = cv2.imread('kucing-photoaidcom-greyscale.jpg', cv2.IMREAD_GRAYSCALE)

# Tambahkan pemeriksaan di sini
if img is None:
    print("Error: Gagal memuat citra. Pastikan nama file benar dan file tidak rusak.")
else:
    # Definisikan Kernel
    lpf_kernel = np.array([[1/9, 1/9, 1/9],
                           [1/9, 1/9, 1/9],
                           [1/9, 1/9, 1/9]])

    hpf_kernel = np.array([[-1, -1, -1],
                           [-1, 8, -1],
                           [-1, -1, -1]])

    bpf_kernel = np.array([[0, -1, 0],
                           [-1, 5, -1],
                           [0, -1, 0]])

    # Langkah 2: Terapkan filter
    lpf_img = cv2.filter2D(img, -1, lpf_kernel)
    hpf_img = cv2.filter2D(img, -1, hpf_kernel)
    bpf_img = cv2.filter2D(img, -1, bpf_kernel)

    # Langkah 3: Tampilkan hasilnya
    plt.figure(figsize=(20, 5))

    plt.subplot(141)
    plt.imshow(img, cmap='gray')
    plt.title('Citra Asli')

    plt.subplot(142)
    plt.imshow(lpf_img, cmap='gray')
    plt.title('Citra yang Difilter Low Pass')

    plt.subplot(143)
    plt.imshow(hpf_img, cmap='gray')
    plt.title('Citra yang Difilter High Pass')

    plt.subplot(144)
    plt.imshow(bpf_img, cmap='gray')
    plt.title('Citra yang Difilter Band Pass')

    plt.show()

    # Hitung nilai piksel rata-rata
    avg_original = np.mean(img)
    avg_lpf = np.mean(lpf_img)
    avg_hpf = np.mean(hpf_img)
    avg_bpf = np.mean(bpf_img)

    print(f"Nilai piksel rata-rata Citra Asli: {avg_original}")
    print(f"Nilai piksel rata-rata Citra LPF: {avg_lpf}")
    print(f"Nilai piksel rata-rata Citra HPF: {avg_hpf}")
    print(f"Nilai piksel rata-rata Citra BPF: {avg_bpf}")
````
